* bruke ````from plotting_utils.plotting_utils import plot_tour````


````plot_tour()```` tar fire argumenter: <br>
&emsp;&emsp;````tour````:<br>
&emsp;&emsp;&emsp;&emsp;&emsp;En liste med navn på byer som skal plottes eller en liste (eller np.array) med bynummere. <br>
&emsp;&emsp;&emsp;&emsp;&emsp;Nummeret på byene tilsvarer nummeret i european_cities.csv-filen. Barcelona er nr. 0, Beograd nr. 1 osv.<br>

&emsp;&emsp;````show_map````:<br>
&emsp;&emsp;&emsp;&emsp;&emsp;Valgfritt. Boolean, default = False. Hvis True, vises plottet med en kartbakgrunn. <br> 
&emsp;&emsp;&emsp;&emsp;&emsp;(Obs: markeringene på kartet passer ikke 100% med prikkene på ruten.)

&emsp;&emsp;````figsize````:<br>
&emsp;&emsp;&emsp;&emsp;&emsp;Valgfritt. Default = (8, 8). Angir størelsen på plottet. 

&emsp;&emsp; ````annotate````:<br>
&emsp;&emsp;&emsp;&emsp;&emsp;Valgfritt. Boolean. Som default skrives det ikke navn ved punktene hvis show_map = True er valgt. <br>&emsp;&emsp;&emsp;&emsp;&emsp;Med annotate = True skrivesallikevel navn på byene på kartet. (Dette blir ikke så veldig pent.)

`````python
#Her er noen eksempler på bruk av plotting-funksjonen:
`````

`````python
tour = ["Madrid", "barcelona", "BERlIN", "Paris", "Rome"]
plot_tour(tour, figsize = (7, 7))
`````

`````python
plot_tour([3, 6, 11, 7], show_map = True)
`````

`````python
plot_tour(["madrid", "Belgrade", "Paris"], show_map = True, figsize = (12, 8), annotate = True)
`````

`````python
tour = np.random.choice(np.arange(20), 6, replace = False)
print("Tour = ", tour)

plot_tour(tour, show_map = True, figsize = (12, 8))
`````
